## PCA Analysis

PCA analysis is done to assess any pattern in the data so as to deploy the best strategy for classification.

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
p=PCA(n_components=2)

scaled_data=StandardScaler().fit_transform(df[df!=-999].iloc[:,1:-2].dropna())

x_pca=p.fit_transform(scaled_data)
```

```python
print('Explained variance ratios of the Principal Axes: '+str(p.explained_variance_ratio_))
```

The variance captured by the principal component axes are not very good, but might shed some light on the composition of events

```python
sns.scatterplot(x=x_pca[:,0],y=x_pca[:,1],hue=df[df!=-999].dropna().Label,edgecolors=None)
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.show()
```

The initial analysis suggest a clustering algorithm like KNN or ensemble classification methods might yield results but the latter is preferred since the former is computationally taxing for 250,000 data points.

## Analysing Missing Values

 Lets analyse the missing values and come up with strategies to deal with them.

As mentioned earlier, the missing values are encoded as -999

```python
df[df.DER_mass_MMC==-999]
```

Let us visualy analyse the missing values with the help of seaborn's `heatmap`

```python
plt.figure(figsize=(10,6))
sns.heatmap(df==-999,cmap='viridis')
plt.yticks([])
plt.show()
```

- The missing values appear to be more frequent in columns which describe the jet features which are essential pseudo particle streams
- There appears the missing values in `jet_leading` columns are a subset of the missing values in `jet_subleading`,`jet_jet` and `DER_lep_eta_centrality`
- There also seems to be some missing values in `DER_mass_MMC` unrelated to the the columns

### Lets try to figure out how the missing values are dependent on the number of jets

```python
df.PRI_jet_num.unique()
```

The proportion of data points against jet numbers:

```python
sns.countplot(x=df.PRI_jet_num)
plt.show()
```

```python
fig,axes=plt.subplots(1,2,figsize=(10,5))
sns.heatmap(df[df.PRI_jet_num==0]==-999,cmap='viridis',ax=axes[0],cbar=False)
axes[0].set_yticks([])
sns.heatmap(df[df.PRI_jet_num==1]==-999,cmap='viridis',ax=axes[1],cbar=False)
axes[1].set_yticks([])
plt.show()
```

- The `leading` and `subleading` values are missing when `PRI_jet_num=0`
- Only the `subleading` values are missing when `PRI_jet_num=1`
- This makes sense, since the `jet` values can only be objserved if there are jets in the experiment and subleading jets does bot make any sense without the presence of a leading jet ie, there must at least be one jet
- The missing values are infact undefined values and it would not make much sense to impute them, therefore we could leave it as an outrageous value like -999 or change it to `NaN` and use an algorithm like XGBoost.

#### Missing Mass Values

```python
plt.figure(figsize=(10,6))
sns.heatmap(df[df.DER_mass_MMC==-999]==-999,cmap='viridis')
plt.yticks([])
plt.show()
```

- Missing mass values are not related to any other missing features

```python
sns.histplot(x=df[df!=-999].DER_mass_MMC)
plt.show()
```

Lets have a look at how correlated is `DER_mass_MMC` to the other features present in the dataset

```python
df[df!=-999].corr(numeric_only=True).DER_mass_MMC.sort_values()[:-1].plot.bar()
plt.title('Correlation DER_mass_MMC ')
plt.ylabel('corr')
plt.show()
```

- `DER_mass_MMC` seems to be highly correlated with `DER_mass_vis` and `DER_deltar_tau_lep`!

```python
fig,axes=plt.subplots(1,2,sharex=True,figsize=(10,5))
sns.scatterplot(df[df!=-999],x='DER_mass_MMC',y='DER_mass_vis',ax=axes[0])
sns.scatterplot(df[df!=-999],x='DER_mass_MMC',y='DER_deltar_tau_lep',ax=axes[1])
plt.show()
```

- The correlated features appears to display a linear trend with respect to `DER_mass_MMC`
- This provides us with an oppurtunity to impute the missing values using a **Linear Regression** or something non linear or non parametric like **KNN** and **Random Forest Algorithm**

### Further Analysis

Further analysis is done to study the dependence of all the features on the target 

```python
for column in df.columns[1:-2]:
    plt.figure()
    sns.histplot(x=df[df!=-999][column],hue=df.Label)
    plt.show()
```

- The $\phi$ `PHI` derived variables excpet `DER_met_phi_centrailty`,`PRI_tau_eta` and `PRI_lep_eta` does not appear to influence the target 
- These columns could be dropped without much loss in information and could potentially reduce overfitting
- It would be pragmatic to drop `jet_Phi` coluns since they already contain too many missing values

Analysing whether the Phi features influence any other features present in the dataset:

```python
phi_cols=[cols for cols in df.columns if 'phi' in cols.split('_')][1:]

for cols in phi_cols:
    plt.figure()
    plt.title(cols)
    df[df!=-999][df.columns[1:-2]].corr()[cols].sort_values()[:-1].plot.bar()
```

There doesnt appear to be a strong correlation between these and other features but are somewhat correlated with other phi values

Let's look at a pair plot between these phi values for good measure

```python
sns.pairplot(data=df[df!=-999][phi_cols])
plt.show()
```
